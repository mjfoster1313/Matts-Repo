```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/tidyverse/ggplot2/main/data-raw/diamonds.csv'
df =  pd.read_csv(url)
```

```python
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
numerical_stats = df[numerical_columns].describe()
display(numerical_stats)
```

```python
categorical_columns = ['cut', 'color', 'clarity']
categorical_stats = df[categorical_columns].describe()
display(categorical_stats)
```

```python
print("Unique columns for Cut: ", df['cut'].unique())
print("Unique columns for Color: ", df['color'].unique())
print("Unique columns for Clarity: ", df['clarity'].unique())
```
